```python
X_train_filtered = X_train.filter(best_params.loc[0:28, 'feature'].values)
X_train_filtered['area_height'] = X_train_filtered['area_percentage'] / X_train_filtered['height_percentage']
X_train_filtered['height_per_floor'] = X_train_filtered['height_percentage']/X_train_filtered['count_floors_pre_eq']
X_train_filtered['volume'] = X_train_filtered['height_percentage']*X_train_filtered['area_percentage']
```

```python
full = X_train_filtered.copy()
full['damage_grade'] = train_labels['damage_grade']
mean_geo_3 = full.groupby(['geo_level_3_id'])['damage_grade'].mean().to_dict()
mean_geo_2 = full.groupby(['geo_level_2_id'])['damage_grade'].mean().to_dict()
mean_geo_1 = full.groupby(['geo_level_1_id'])['damage_grade'].mean().to_dict()

X_train_filtered['geo_level_1_id'] = X_train_filtered['geo_level_1_id'].map(mean_geo_1)
X_train_filtered['geo_level_2_id'] = X_train_filtered['geo_level_2_id'].map(mean_geo_2)
X_train_filtered['geo_level_3_id'] = X_train_filtered['geo_level_3_id'].map(mean_geo_3)
X_train_filtered['geo_level_3_id'] = X_train_filtered['geo_level_3_id']*X_train_filtered['geo_level_2_id']

X_train_filtered
```

```python
X_test_filtered['geo_level_1_id'] = X_test_filtered['geo_level_1_id'].map(mean_geo_1)
X_test_filtered['geo_level_2_id'] = X_test_filtered['geo_level_2_id'].map(mean_geo_2)
X_test_filtered['geo_level_3_id'] = X_test_filtered['geo_level_3_id'].map(mean_geo_3)
X_test_filtered['geo_level_3_id'] = X_test_filtered['geo_level_3_id']*X_test_filtered['geo_level_2_id']

X_test_filtered
```
